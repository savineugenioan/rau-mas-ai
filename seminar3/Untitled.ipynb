### This is our first ML tutorial. 

In this tutorial we are going to cover a basic linear regression algorithm. 

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
```

```python
diabetes = load_diabetes()
X, y = diabetes.data, diabetes.target
print(f"Primul sample: {X[0]}")
print(f"Primul output: {y[0]}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
print(f"Full dataset: {len(X)}")
print(f"Train dataset: {len(X_train)}")
print(f"Test dataset: {len(X_test)}")
```

```python
from sklearn.linear_model import LinearRegression, Ridge

# y = w0*x0 + w1*x1 + ... + wn*xn, x0 = 1, x1, ... xn = variabilele noastre (coloanele din setul de date)
# x1 = cos(x), xn = x**10
# r = LinearRegression()
r = Ridge(alpha=0.1)
r = r.fit(X_train, y_train)
```

```python
print(f"Intercept: {r.intercept_}")
print(f"Coefficients: {r.coef_}")
```

```python
from sklearn.metrics import mean_squared_error

y_pred_train = r.predict(X_train)
y_pred_test = r.predict(X_test)

mse_train = mean_squared_error(y_pred=y_pred_train, y_true=y_train)
mse_test = mean_squared_error(y_pred=y_pred_test, y_true=y_test)

print(f"Training MSE: {mse_train}")
print(f"Test MSE: {mse_test}")
```

```python
from sklearn.metrics import mean_absolute_error

y_pred_train = r.predict(X_train)
y_pred_test = r.predict(X_test)

mae_train = mean_absolute_error(y_pred=y_pred_train, y_true=y_train)
mae_test = mean_absolute_error(y_pred=y_pred_test, y_true=y_test)

print(f"Training MAE: {mae_train}")
print(f"Test MAE: {mae_test}")
```

```python
import matplotlib.pyplot as plt
from numpy import abs
%matplotlib inline 

_, ax = plt.subplots(2, figsize=(20, 10))
ax[0].bar(range(len(y_train)), abs(y_pred_train-y_train))
ax[1].bar(range(len(y_test)), abs(y_pred_test-y_test))
plt.show()
```

```python
from sklearn.linear_model import Lasso

# y = w0*x0 + w1*x1 + ... + wn*xn, x0 = 1, x1, ... xn = variabilele noastre (coloanele din setul de date)
# r = LinearRegression()
r = Lasso()
r = r.fit(X_train, y_train)
```

```python
y_pred_train = r.predict(X_train)
y_pred_test = r.predict(X_test)

mse_train = mean_squared_error(y_pred=y_pred_train, y_true=y_train)
mse_test = mean_squared_error(y_pred=y_pred_test, y_true=y_test)

print(f"Training MSE: {mse_train}")
print(f"Test MSE: {mse_test}")
```
