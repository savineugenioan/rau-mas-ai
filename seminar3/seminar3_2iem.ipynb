```python
from sklearn.datasets import load_diabetes
diabetes_data = load_diabetes()
X, y = diabetes_data.data, diabetes_data.target
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
run = False
if run:
    n_vars = X.shape[1]
    for i in range(n_vars):
        _ = plt.figure(figsize=(20, 10))
        plt.hist(X[:,i], bins=30)
        plt.title = f"Column {i}"
        plt.show()
X = X[:, 1:]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge

r = Lasso()
r = r.fit(X_train, y_train)
```

```python
r.coef_
```

```python
r.intercept_
```

```python
y_train_pred = r.predict(X_train)
y_test_pred = r.predict(X_test)
```

```python
import numpy as np
y_train_diff = np.abs(y_train_pred - y_train)
y_test_diff = np.abs(y_test_pred - y_test)

n_examples_train = range(len(y_train_pred))
n_examples_test = range(len(y_test_pred))
_, axs = plt.subplots(2, figsize=(20, 10))
axs[0].bar(n_examples_train, y_train_diff)
axs[1].bar(n_examples_test, y_test_diff)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mse_train = mean_squared_error(y_true=y_train, y_pred=y_train_pred)
mse_test = mean_squared_error(y_true=y_test, y_pred=y_test_pred)
print(f"MSE Train: {mse_train}")
print(f"MSE Test: {mse_test}")
```
